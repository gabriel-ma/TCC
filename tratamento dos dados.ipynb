```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.set_option('display.max_columns', 500)
movies = pd.read_csv('movie_metadata_0.csv')
movies.drop_duplicates(inplace=True)
movies.info()
```

```python
movies.budget.fillna(movies.budget.mean(), inplace=True)
movies.title_year.fillna(movies.title_year.mean(), inplace=True)
movies.num_critic_for_reviews.fillna(movies.num_critic_for_reviews.mean(), inplace=True)
movies.duration.fillna(movies.duration.mean(), inplace=True)
movies.actor_3_facebook_likes.fillna(movies.actor_3_facebook_likes.mean(), inplace=True)
movies.actor_1_facebook_likes.fillna(movies.actor_1_facebook_likes.mean(), inplace=True)
movies.actor_2_facebook_likes.fillna(movies.actor_2_facebook_likes.mean(), inplace=True)
movies.facenumber_in_poster.fillna(movies.facenumber_in_poster.mean(), inplace=True)
movies.num_user_for_reviews.fillna(movies.num_user_for_reviews.mean(), inplace=True)
movies.gross.fillna(0, inplace=True)
```

```python
movies.num_critic_for_reviews = movies.num_critic_for_reviews.astype('int64')
movies.director_facebook_likes = movies.director_facebook_likes.astype('int64')
movies.duration = movies.duration.astype('int64')
movies.gross = movies.gross.astype('int64')
movies.actor_3_facebook_likes = movies.actor_3_facebook_likes.astype('int64')
movies.actor_1_facebook_likes = movies.actor_1_facebook_likes.astype('int64')
movies.actor_2_facebook_likes = movies.actor_2_facebook_likes.astype('int64')
movies.facenumber_in_poster = movies.facenumber_in_poster.astype('int64')
movies.num_user_for_reviews = movies.num_user_for_reviews.astype('int64')
movies.title_year = movies.title_year.astype('int64')
```

```python
movies.to_csv('movies_metadata_cleaning1.csv', index=False)
```
